```python
#Q#3
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
data = {
    'Height (cm)': [160, 175, 168, 182, 155, 170, 180, 165, 172, 158],
    'Weight (kg)': [55, 75, 65, 80, 50, 68, 77, 62, 72, 54]
}

data = pd.DataFrame(data)

print(data)


data.plot(x='Height (cm)', y='Weight (kg)', kind='scatter')


plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid() 

plt.show()
```

```python
#QUESTION_06
import pandas as pd

df = pd.read_csv("C:/Users/fa22-bscs-0205/Desktop/income - Copy.csv")
df
```

```python
income_threshold = 77  
age_threshold = 30       

df['IncomeAboveThreshold'] = (df['Income'] > income_threshold) & (df['Age'] < age_threshold)

print("Original DataFrame with IncomeAboveThreshold:")
print(df)

new_dataframe = df[df['IncomeAboveThreshold']][['ID', 'Age', 'Income']].reset_index()

print("\nNew DataFrame for Individuals Above Threshold:")
print(new_dataframe)
```

```python
mean_age_filtered = new_dataframe['Age'].mean()
mean_income_filtered = new_dataframe['Income'].mean()

mean_age_unfiltered = df['Age'].mean()
mean_income_unfiltered = df['Income'].mean()

print("\nMean Age (Filtered):", mean_age_filtered)
print("Mean Income (Filtered):", mean_income_filtered)
print("Mean Age (Unfiltered):", mean_age_unfiltered)
print("Mean Income (Unfiltered):", mean_income_unfiltered)

categories = ['Filtered', 'Unfiltered']
mean_incomes = [mean_income_filtered, mean_income_unfiltered]

plt.bar(categories, mean_incomes, color=['skyblue', 'lightgreen'])
plt.ylabel('Mean Income')
plt.title('Mean Income Comparison: Filtered vs. Unfiltered')
plt.grid(axis='y')

plt.show()
```

```python
#Q#10
daily_temperatures = np.random.randint(11, 44 + 1, 100)

print("Daily Temperatures for 100 Days:")
print(daily_temperatures)
```

```python
df = pd.DataFrame({
    'Day': np.arange(1, 100 + 1),  # Days from 1 to 100
    'Temperature': daily_temperatures
})

df['7-Day Rolling Average'] = df['Temperature'].rolling(window=7).mean()

print("Daily Temperatures and 7-Day Rolling Average:")
print(df)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Day'], df['Temperature'], marker='*', label='Daily Temperature', color='skyblue')
plt.plot(df['Day'], df['7-Day Rolling Average'], label='7-Day Rolling Average', color='orange', linewidth=2)
plt.title('Daily Temperatures with 7-Day Rolling Average')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.xticks(np.arange(0, 100 + 1, 10)) 
plt.grid()
plt.show()
```

```python
#Q#9
import pandas as pd
import numpy as np


num_products = 100  
product_ids = np.arange(1, num_products + 1) 
prices = np.round(np.random.uniform(10, 500, size=num_products), 2) 
quantities_sold = np.random.randint(1, 100, size=num_products)  


ecommerce_data = pd.DataFrame({
    'Product ID': product_ids,
    'Price': prices,
    'Quantity Sold': quantities_sold
})

print(ecommerce_data.head())
```

```python
•	Create a random NumPy array representing stock prices for 30 days.
•	Store the prices in a Pandas DataFrame with columns for “Day” and “Price.”
•	Calculate the percentage change in stock price from one day to the next and store it in a new column.
•	Plot a line graph to show the daily stock price and a separate plot showing the percentage change over time.
```

```python
#Q#7
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

days= 30

stock = np.random.randint(20000, 770000, days)
days = np.arange(1, 31)

d= {
    'Days': days,
    'Stock': stock
}
d = pd.DataFrame(d)

print(d)
```

```python
d['Percentage_Change'] = d['Stock'].pct_change() * 100
print(d)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].plot(d['Days'], d['Stock'], color='blue', marker='o', linestyle='-', label='Stock Price')
ax[0].set_title('Daily Stock Price')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Stock')
ax[0].legend()

ax[1].plot(d['Days'], d['Stock'], color='red', marker='o', linestyle='-', label='Percentage Change')
ax[1].set_title('Percentage Change in Stock Price')
ax[1].set_xlabel('Days')
ax[1].set_ylabel('Stock')
ax[1].legend()

plt.tight_layout()

# Show the plots
plt.show()
```
