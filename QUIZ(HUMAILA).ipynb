```python
#Q#3
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,80,1000)
d_r=1*t
d_s=4*(t-30)
fig, ax=plt.subplots()
plt.title("Jills Solar panels")
plt.xlabel("days")
plt.ylabel("energy in joules")
ax.set_xlim([0,100])
ax.set_ylim([0,100])
ax.plot(t,d_r,color='green')
ax.plot(t,d_s,color='red')
plt.axvline(x=40,color='purple',linestyle="--")
plt.axhline(y=40,color='blue',linestyle="--")
```

```python
#Q5

import numpy as np
import tensorflow as tf
import torch

A = np.array([[4, 7, 2],
              [3, 5, 1],
              [2, 3, 1]])


i = np.array(A)
x_inv = np.linalg.inv(i)
print('inverse of x=',x_inv)

A_tf = tf.constant(A, dtype=tf.float32)
intensor = tf.linalg.inv(A_tf)
print("Inverse of A using TensorFlow:\n", intensor.numpy())


A_pt = torch.tensor(A, dtype=torch.float32)
pytorch = torch.inverse(A_pt)
print("Inverse of A using PyTorch:\n", pytorch)
```
